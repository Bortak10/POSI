## Introduction

### Naive Bayes Algorithm

The Naive Bayes algorithm is a classic probabilistic algorithm used primarily for classification. It is a model based on Bayes’ theorem with the assumption of feature independence. This means that for a given classified object, individual features (variables) are treated as independent, which significantly simplifies the computations.

## Bayes' Theorem

Bayes’ theorem describes the relationship between conditional, prior, and posterior probabilities:

$$
P(C \mid X) = \frac{P(X \mid C) P(C)}{P(X)}
$$

where:

- $ P(C \mid X) $ — the probability that an object belongs to class $ C $, given features $ X $ (posterior),
- $ P(X \mid C) $ — the probability of observing features $ X $ if the object belongs to class $ C $ (likelihood),
- $ P(C) $ — the prior probability that an object belongs to class $ C $,
- $ P(X) $ — the probability of observing features $ X $ (often treated as a constant, ignored in classification).

<br>

## Assumptions of Naive Bayes

Naive Bayes assumes that the features are independent. This means that each feature is treated separately in the calculations, ignoring mutual dependencies between them. Thanks to this, the algorithm is very fast and efficient, although it does not always achieve the best results compared to more advanced models.

## Modeling in Naive Bayes

To compute the probability that an object belongs to class $ C $, we use the following formula:

$$
P(C \mid X) = P(C) \prod_{i=1}^{n} P(x_i \mid C) \frac{1}{P(X)}
$$

$$
\prod_{i=1}^{n} P(x_i \mid C) = P(x_1 \mid C) \cdot P(x_2 \mid C) \cdot \cdots \cdot P(x_i \mid C)
$$

$$
P(C \mid X) = \frac {P(C) \cdot P(x_1 \mid C) \cdot P(x_2 \mid C) \cdot \cdots \cdot P(x_i \mid C)}{P(X)}
$$

<br>

We may also ignore the denominator, because it is the same for all classes and does not affect the classification. Then:

$$
P(C \mid X) \propto P(C) \cdot P(x_1 \mid C) \cdot P(x_2 \mid C) \cdot \cdots \cdot P(x_i \mid C)
$$

where:

- $ x_1, x_2, \dots, x_n $ are the features of the object $ X $,
- $ P(x_i \mid C) $ is the probability of observing feature $ x_i $ given that the object belongs to class $ C $.

In practice, the algorithm computes probabilities for each class and selects the class with the highest probability.

### Step by step:

1. Compute the prior probability $ P(C) $ for each class $ C $, i.e., the probability that a given object belongs to class $ C $:

$$
P(C) = \frac{\text{number of samples in class } C}{\text{total number of samples}}
$$

2. Compute the conditional probabilities $ P(x_i \mid C) $ for each feature $ x_i $ and class $ C $. This can be done based on the frequency of the feature $ x_i $ in class $ C $.

3. Apply Bayes’ formula for each class $ C $ and choose the class for which $ P(C \mid X) $ is the highest.

### Calculating Probabilities Using Bayes' Formula

## 1. Compute the prior probabilities $ P(A) $ and $ P(B) $

We have 5 samples in the dataset, of which:
- 3 samples belong to class $ A $ (in the first, second, and fourth rows),
- 2 samples belong to class $ B $ (in the third and fifth rows).

We compute the prior probabilities for each class:

$$
P(A) = \frac{3}{5} = 0.6
$$

$$
P(B) = \frac{2}{5} = 0.4
$$

<br>

## 2. Compute the conditional probabilities $ P(x_1 \mid A) $, $ P(x_1 \mid B) $, $ P(x_2 \mid A) $, and $ P(x_2 \mid B) $

### Feature 1 ($ x_1 $):

$$
P(x_1 = \text{Yes} \mid A) \quad \text{— How many times is feature 1 equal to "Yes" in class } A?
$$

In class $ A $ we have 3 samples, of which 2 have feature 1 equal to "Yes" (the first and second samples). Thus:

$$
P(x_1 = \text{Yes} \mid A) = \frac{2}{3}
$$

<br>

$$
P(x_1 = \text{Yes} \mid B) \quad \text{— How many times is feature 1 equal to "Yes" in class } B?
$$

In class $ B $ we have 2 samples, of which 1 has feature 1 equal to "Yes" (the third sample). Thus:

$$
P(x_1 = \text{Yes} \mid B) = \frac{1}{2}
$$

<br>
<br>

### Feature 2 ($ x_2 $):

$$
P(x_2 = \text{Yes} \mid A) \quad \text{— How many times is feature 2 equal to "Yes" in class } A?
$$

In class $ A $ we have 3 samples, of which 2 have feature 2 equal to "Yes" (the first and fourth samples). Thus:

$$
P(x_2 = \text{Yes} \mid A) = \frac{2}{3}
$$

$$
P(x_2 = \text{Yes} \mid B) \quad \text{— How many times is feature 2 equal to "Yes" in class } B?
$$

In class $ B $ we have 2 samples, of which 1 has feature 2 equal to "Yes" (the third sample). Thus:

$$
P(x_2 = \text{Yes} \mid B) = \frac{1}{2}
$$

<br>
<br>

## 3. Compute $ P(A \mid X) $ and $ P(B \mid X) $

We have a new object with the following features:
- Feature 1 = "Yes"
- Feature 2 = "No"

Thus we compute $ P(x_1 = \text{Yes}, x_2 = \text{No}) $.

### Calculating $ P(A \mid X) $:

Let us use Bayes’ formula for class $ A $:

$$
P(A \mid X) \propto P(A) \cdot P(x_1 = \text{Yes} \mid A) \cdot P(x_2 = \text{No} \mid A)
$$

$$
P(x_1 = \text{Yes} \mid A) = \frac{2}{3}, \quad P(x_2 = \text{No} \mid A) = 1 - \frac{2}{3} = \frac{1}{3}
$$

$$
P(A \mid X) \propto 0.6 \cdot \frac{2}{3} \cdot \frac{1}{3} = 0.6 \cdot \frac{2}{9} = \frac{1.2}{9} = \frac{2}{15}
$$

### Calculating $ P(B \mid X) $:

Use Bayes’ formula for class $ B $:

$$
P(B \mid X) \propto P(B) \cdot P(x_1 = \text{Yes} \mid B) \cdot P(x_2 = \text{No} \mid B)
$$

$$
P(x_1 = \text{Yes} \mid B) = \frac{1}{2}, \quad P(x_2 = \text{No} \mid B) = 1 - \frac{1}{2} = \frac{1}{2}
$$

$$
P(B \mid X) \propto 0.4 \cdot \frac{1}{2} \cdot \frac{1}{2} = 0.4 \cdot \frac{1}{4} = \frac{0.4}{4} = \frac{1}{10}
$$

## 4. Class Selection

Comparing the values of $ P(A \mid X) $ and $ P(B \mid X) $, we get:

$$
P(A \mid X) = \frac{2}{15}, \quad P(B \mid X) = \frac{1}{10} = \frac{1.5}{15}
$$

Since $ P(B \mid X) > P(A \mid X) $, the object will be assigned to class $ B $.


### Example of Using the Bayes Classifier in Python Code

````python
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, classification_report

iris = load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = GaussianNB()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")
print(classification_report(y_test, y_pred))
````
