```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import missingno as msno

from google.colab import files
files.upload()

df = pd.read_csv('titanic.csv')


duplicates = df.duplicated(subset=["PassengerId"])
print("Number of duplicates:",duplicates.sum())
print()
print("Basic information about the dataset")
display(df.info())
print()

df["Age"] = df["Age"].fillna(df["Age"].median())
df["Sex"] = df["Sex"].astype("category")
df["Cabin"] = df["Cabin"].fillna("N/A")
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode())

print("Table of null values after update")
display(df.isnull().sum())
print()

print("First rows:")
display(df.head())
print()

print("Boxplot of Age:")
print()
plt.boxplot(df["Age"].dropna())
plt.title("Boxplot of Age")
plt.ylabel("Age")
plt.show()
print()

print("Boxplot of Fares:")
print()
plt.boxplot(df["Fare"].dropna())
plt.title("Boxplot of Fares")
plt.ylabel("Fares")
plt.show()
print()

fare_mean = df["Fare"].mean()
fare_std = df["Fare"].std()

print(f"Mean Fare: {fare_mean:.2f}")
print(f"Standard deviation: {fare_std:.2f}")


df["z_score"] = (df["Fare"] - fare_mean) / fare_std


outliers_z = df[np.abs(df["z_score"]) > 3]

print("\nðŸ” Outliers Detected (|Z| > 3):")
print()
print(outliers_z)



print()
print("Table of the statistical data for all numerical columns:")
display(df.describe())
print()
print("Survivors = 1, Fatalities = 0")
display(df['Survived'].value_counts())
print()
print("Sex of the passengers:")
display(df['Sex'].value_counts())
print()
print("Port of Embarkation information:")
display(df['Embarked'].value_counts())
print()
print("Passenger Classes information:")
display(df['Pclass'].value_counts())
print()

print("Charts displaying: Age, Sibilings and Spouses, Parents and Children, Fares, Passenger Classes, Sex of the Passengers, Ports of Embarkation and Comparison of Age vs Passenger Class with Survival")
fig, axes = plt.subplots(2, 4, figsize=(20, 8))

axes[0,0].hist(df['Age'],bins=20, color='darkblue', edgecolor='black')
axes[0, 0].set_title("Age Distribution")

axes[0,1].hist(df['SibSp'],bins=8, color='lightgreen', edgecolor='black')
axes[0, 1].set_title("Sibilings and Spouses Distribution")

axes[0,2].hist(df['Parch'],bins=6, color='darkorange', edgecolor='black')
axes[0, 2].set_title("Parents and Children Distribution")

axes[0,3].hist(df['Fare'],bins=20, color='darkred', edgecolor='black')
axes[0, 3].set_title("Fare Distribution")

df["Pclass"].value_counts().plot(kind='pie', ax=axes[1, 0], autopct='%1.1f%%',
                startangle=90, colors=sns.color_palette("pastel"), ylabel='')
axes[1, 0].set_title("Passenger Classes")

df["Sex"].value_counts().plot(kind='pie', ax=axes[1, 1], autopct='%1.1f%%',
                startangle=90, colors=sns.color_palette("pastel"), ylabel='')
axes[1, 1].set_title("Sex of the Passengers")

df["Embarked"].value_counts().plot(kind='pie', ax=axes[1, 2], autopct='%1.1f%%',
                startangle=90, colors=sns.color_palette("pastel"), ylabel='')
axes[1, 2].set_title("Ports of Embarkation")

df["Survived"].value_counts().plot(kind='pie', ax=axes[1, 3], autopct='%1.1f%%',
                startangle=90, colors=sns.color_palette("pastel"), ylabel='')
axes[1, 3].set_title("Survival information")

print()
print()
plt.figure(figsize=(7,5))
sns.scatterplot(x="Age", y="Pclass", hue="Survived", data=df,)
plt.title("(Scatter Plot: Age vs Passenger Class)")
plt.show()

print()
print("Age vs fare")
print()
plt.figure(figsize=(7,5))
sns.scatterplot(x="Age", y="Fare", hue="Survived", data=df,)
plt.title("Scatter Plot: Age vs Fare)")
plt.show()

print()
print("Age vs Sex")
print()
plt.figure(figsize=(7,5))
sns.scatterplot(x="Age", y="Survived", hue="Sex", data=df,)
plt.title("Scatter Plot: Sex vs Age)")
plt.show()

print()
print("Age vs Sibilings & Spouses")
print()
plt.figure(figsize=(7,5))
sns.scatterplot(x="Age", y="SibSp", hue="Survived", data=df,)
plt.title("Scatter Plot: Sibilings or Spouses vs Age)")
plt.show()

print()
print("Age vs Embarkation")
print()
plt.figure(figsize=(7,5))
sns.scatterplot(x="Age", y="Embarked", hue="Survived", data=df,)
plt.title("Scatter Plot: Embarkation vs Age)")
plt.show()


men_survived = df.loc[df['Sex'] == 'male', 'Survived'].sum()
women_survived = df.loc[df['Sex'] == 'female', 'Survived'].sum()


sex = ['Men', 'Women']
Survivors = [men_survived, women_survived]

print()
print("Bar chart of Survival splitted by Sex")
print()
plt.bar(sex, Survivors, color='violet')
plt.title('Survival and Sex Correlation')
plt.xlabel('Survivors')
plt.ylabel('Sex')
plt.show()

first_survival = df.loc[df['Pclass'] == 1, 'Survived'].sum()
second_survived = df.loc[df['Pclass'] == 2, 'Survived'].sum()
third_survived = df.loc[df['Pclass'] == 3, 'Survived'].sum()

passclass = ['1stClass', '2ndClass', '3rdClass']
passSurvivors = [first_survival, second_survived,third_survived]

print()
print("Bar chart of Survival splitted by Passenger Class")
print()
plt.bar(passclass, passSurvivors, color='green')
plt.title('Passenger Class and Survival Correlation')
plt.xlabel('Survivors')
plt.ylabel('Passenger Class')
plt.show()

southampton = df.loc[df['Embarked'] == 'S', 'Survived'].sum()
cherbourg = df.loc[df['Embarked'] == 'C', 'Survived'].sum()
queenstown = df.loc[df['Embarked'] == 'Q', 'Survived'].sum()
na = df.loc[df['Embarked'] == 'N/A', 'Survived'].sum()

port = ['Southampton', 'Cherbourg', 'Queenstown', 'N/A']
portSurvivors = [southampton, cherbourg ,queenstown, na]

print()
print("Bar chart of Survival splitted by Port of embarkation")
print()
plt.bar(port, portSurvivors, color='yellow')
plt.title('Port of embarkation and Survival Correlation')
plt.xlabel('Survivors')
plt.ylabel('Port')
plt.show()

print()
print("Heatmap of Correlation")
print()
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap")
plt.show()



```
