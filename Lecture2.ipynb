# Lecture 2 - Exploratory Data Analysis (EDA)

## What is EDA and why it’s important?
Imagine you’re a manager stepping into a new company.
Before making any decisions — like changing the strategy, hiring people, or launching a product — you’d first want to understand how things currently look:
- What are the sales numbers?
- Which regions are performing well?
- Which products are losing money?

That first step of exploring and understanding before acting — that’s exactly what Exploratory Data Analysis (EDA) is in the world of data.

EDA is the process of looking carefully at your data to understand what’s inside before building any predictive models or making business decisions.
It’s like doing a "check-up" on your data’s health — you want to know what you can trust, what’s missing, and what stands out.

## Understanding the data before modeling
Think of building a data model like cooking a meal.
Before you start mixing ingredients, you first check:
- What ingredients do I have?
- Are they fresh or spoiled?
- Do I have enough of each one?

In the same way, before we create a statistical or machine learning model, we need to inspect the data:
- Are there missing values?
- Are all columns in the correct format?
- Are some numbers extremely high or low (outliers)?

If we skip this step, we risk “cooking” a model that gives wrong or misleading results.

## Detecting patterns, trends, anomalies, and relationships
EDA helps us see what’s happening inside the data — much like how a good manager reads between the lines in a financial report.

For example:
- A pattern might be that sales increase every December.
- A trend might be that customer satisfaction is slowly declining over months.
- An anomaly (something unusual) might be that one store sells five times more than others — which could be a success story or a data entry mistake.
- A relationship might be that higher marketing spending leads to higher sales — but only in certain regions.
- These insights help us ask better business questions and prepare for deeper analysis.

## Typical workflow in a data science project
EDA is just one part of a bigger data science/AI process.
Here’s how it usually goes:

1. Data collection – gather information (from databases, surveys, sensors, etc.)
2. Data cleaning – remove errors, fill in missing values, fix data types
3. Exploration (EDA) – understand the data using visualizations and statistics
4. Modeling – use machine learning or statistical methods to make predictions or classifications
5. Evaluation – test how well the model works and if it makes sense for business decisions

Think of this like preparing a new market entry strategy:
1. You collect market data,
2. Clean and verify it,
3. Explore it to understand consumer behavior,
4. Build a model or forecast,
5. And finally evaluate if your strategy works.

## Goals of EDA
EDA has several important goals:

Form hypotheses
→ Like a manager forming questions based on data:
“Why do customers in Region A spend more than in Region B?”
“Is the drop in sales caused by seasonality or competition?”

Guide feature engineering
→ This means deciding which variables are most useful for our models.
For instance, if we notice that “number of visits per month” is highly correlated with “total spend,” we might create a new variable combining them.

Inform model selection
→ Different types of models work better with different kinds of data.
By exploring the data, we get clues about which model (linear regression, decision trees, etc.) might make sense.

## In short
EDA is like a first meeting with your data.
You don’t jump into conclusions — you listen, observe, and ask questions.
It’s a mix of detective work, curiosity, and business thinking.

You use statistics and visualizations to:
- Understand the “story” your data is telling,
- Discover opportunities or problems hidden in the numbers, and
- Prepare for more advanced analysis or predictive modeling.

## Handling Missing Data
### Detecting Missing Data
Missing data can appear as NaN, empty strings, or even incorrectly formatted values.


This gives us the total number of missing entries per column.

Interpretation:

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the CSV file from Google Drive
df = pd.read_csv('/content/drive/MyDrive/POSI/Lecture 2/Data/customers_with_missing.csv')

# Replace common invalid markers with NaN
df = df.replace(["", "N/A", "unknown", "invalid"], np.nan)

print("🔍 Checking missing values in each column:")
print(df.isnull().sum())

# Visualize missingness
plt.imshow(df.isnull(), aspect='auto', interpolation='nearest', cmap="Greys")
plt.title("Missing Values Heatmap (True = missing)")
plt.xlabel("Columns")
plt.ylabel("Rows")
plt.xticks(ticks=range(len(df.columns)), labels=df.columns, rotation=45, ha='right')
plt.colorbar(label="Missing (1=True, 0=False)")
plt.show()
````

Interpretation

- The heatmap visually displays missing cells — bright/white areas are missing data.

This dataset intentionally includes:
- Numeric missing values (None, np.nan)
- Empty strings
- Invalid text placeholders ("N/A", "unknown", "invalid")
- Mixed date formats

````python
import pandas as pd
import numpy as np

# Replace placeholders with NaN
df = df.replace(["", "N/A", "unknown", "invalid"], np.nan)

# Calculate total and percentage of missing values
missing_count = df.isnull().sum()
missing_percentage = (df.isnull().sum() / len(df)) * 100

# Combine both into one summary table
missing_summary = pd.DataFrame({
    "Missing Count": missing_count,
    "Missing %": missing_percentage.round(2)
})

print("🔍 Missing Value Summary (Count and Percentage):")
print(missing_summary)
````

# Dealing with Outliers
Outliers are extreme values that can distort statistics.
We can detect and handle them using visualizations or numeric methods.

## Visual detection — Boxplot

## Boxplot
A boxplot (also known as a box-and-whisker plot) is a graphical summary of the distribution of a dataset.
It helps us quickly spot outliers, check symmetry, and understand spread (variation).

A boxplot shows five key summary statistics:
| Element                 | Meaning                               |
| ----------------------- | ------------------------------------- |
| **Minimum**             | The smallest non-outlier value        |
| **Q1 (First Quartile)** | 25% of the data fall below this value |
| **Median (Q2)**         | The middle value (50%)                |
| **Q3 (Third Quartile)** | 75% of the data fall below this value |
| **Maximum**             | The largest non-outlier value         |

````python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('/content/drive/MyDrive/POSI/Lecture 2/Data/income_with_outliers.csv')

# Create boxplot for income
plt.boxplot(df["income"].dropna(), vert=True)
plt.title("Boxplot of Income")
plt.ylabel("Income")
plt.grid(True)
plt.show()
````

Interpretation:
- The box represents the central 50% of the data (between Q1 and Q3).
- The line inside the box is the median income.
- The dots far away from the box (above the whiskers) are outliers — unusually high incomes.

````python
# Using IQR method

Q1 = df["income"].quantile(0.25)
Q3 = df["income"].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df["income"] < lower_bound) | (df["income"] > upper_bound)]
print(outliers)
````

Interpretation:
- Rows printed here are your outliers — they fall outside the typical range of the data.

The Interquartile Range (IQR) method is a simple and powerful way to detect outliers — values that are much higher or lower than the rest of the data.

Calculate Quartiles
- We divide the sorted data into four equal parts using quartiles:
- Q1 (First Quartile) → 25% of data falls below this value
- Q2 (Median) → 50% of data falls below this value
- Q3 (Third Quartile) → 75% of data falls below this value

The Interquartile Range (IQR) is the distance between Q3 and Q1 — the middle 50% of data.
```
IQR=Q3−Q1
```

Define “Normal” Range
Values that fall too far outside the middle 50% are considered outliers.
We define cutoff boundaries as:
```
Lower bound=Q1−1.5×IQR
Upper bound=Q3+1.5×IQR
```

Any data point below the lower bound or above the upper bound is an outlier.

# Numeric detection — Z-score method
Z-score tells how many standard deviations a value is from the mean.

```
income_mean = df["income"].mean()
income_std = df["income"].std()
df["z_score"] = (df["income"] - income_mean) / income_std

# Flag outliers
outliers_z = df[np.abs(df["z_score"]) > 3]
```

Explanation:
A z-score above 3 or below -3 often indicates an outlier.

```
Z=σ(x−μ)​
```
Where:
- x = individual value
- μ = mean of the dataset
- σ = standard deviation

Interpretation:
- A Z-score close to 0 → value is near the mean
- A Z-score of +3 or -3 → value is far away from the mean
- Values with |Z| > 3 are usually considered outliers

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv('/content/drive/MyDrive/POSI/Lecture 2/Data/income_for_zscore.csv')

# 1️⃣ Calculate mean and standard deviation
income_mean = df["income"].mean()
income_std = df["income"].std()

print(f"Mean income: {income_mean:.2f}")
print(f"Standard deviation: {income_std:.2f}")

# 2️⃣ Compute Z-score for each income
df["z_score"] = (df["income"] - income_mean) / income_std

# 3️⃣ Flag outliers (|Z| > 3)
outliers_z = df[np.abs(df["z_score"]) > 3]

print("\n🔍 Outliers Detected (|Z| > 3):")
print(outliers_z)
````

##Additional explanation

### Average
The mean (often called the average) is the sum of all values divided by the number of values.
| Ages     | 20                                   | 22 | 25 | 30 | 100 |
| -------- | ------------------------------------ | -- | -- | -- | --- |

Mean=(20 + 22 + 25 + 30 + 100) / 5 = 39.4

When to use it:
- When your data is symmetrical (values are evenly spread).
- When there are no extreme outliers (since they distort the mean).
- Good for continuous numeric data, like temperatures or test scores.

Weakness:
- Outliers heavily influence the mean.
- In the example above, the age 100 increases the mean a lot —
the average age looks higher than what’s typical.

### Median
The median is the middle value when all numbers are arranged in order.

If there’s an even number of values, the median is the average of the two middle ones.
| Ages (sorted) | 20                | 22 | 25 | 30 | 100 |
| ------------- | ----------------- | -- | -- | -- | --- |

Median=25 (middle value)

When to use it:
- When the data is skewed (not symmetrical).
- When there are outliers, because the median is not affected by them.
- Good for describing typical values in income, property prices, or wait times.

Weakness:
- Doesn’t reflect the magnitude of values above or below it.
(You lose some information about the spread of data.)

### Mode
The mode is the most frequent value in the dataset.
| Cities   | Warsaw                   | Warsaw | Krakow | Gdansk | Warsaw |
| -------- | ------------------------ | ------ | ------ | ------ | ------ |

Mode=Warsaw (appear 3 times)

When to use it:
- For categorical data (like “city,” “color,” or “product type”).
- For discrete data where numbers repeat often.
- Can describe the most typical or popular category.

Weakness:
- Not always unique (data can be bimodal or multimodal).
- Not useful for data where every value is unique.

## Dispersion (Variability)
These statistics describe how spread out the values are.

```

std_income = df["income"].std()

max_income = df["income"].max()
range_income = max_income - min_income

print(f"Variance: {variance_income:.2f}")
print(f"Standard deviation: {std_income:.2f}")
print(f"Range: {range_income:.2f}")
```

Interpretation:
- Variance and standard deviation measure how far values deviate from the mean.
- A large std means data is more spread out.
- Range gives a simple measure of total spread.

##Additional explanation
Even if two datasets have the same mean, they can be completely different in how spread out their values are.

Dispersion measures how much the data varies around the mean (or median).
It answers the question:

“Are all values close to each other, or are they far apart?”

### Why Dispersion Is Important

It measures consistency.
- Small dispersion → data is tightly grouped → consistent behavior.
- Large dispersion → data is spread out → unpredictable or unstable behavior.

It helps compare datasets.
- Two groups may have the same mean income, but one group may have higher inequality.

It’s essential in statistics and machine learning.
- Many models (like regression) assume data is normally distributed.
- Standard deviation and variance are key in understanding uncertainty, confidence intervals, and risk.
- It reveals hidden patterns.
- Without looking at spread, averages can be misleading.

Interpretation:
- Larger variance → values differ more from the mean.

### Standard Deviation
- Same units as the original data (e.g., $).
- Easier to interpret than variance.
- Tells, on average, how far each value is from the mean.

```
std_income = df["income"].std()
print(f"Standard deviation: {std_income:.2f}")
```

Interpretation:
- If std_income = 10,000, most incomes are about $10,000 away from the average income.

## Summary
| Measure                | What It Shows               | Units        | Pros              | Cons                 |
| ---------------------- | --------------------------- | ------------ | ----------------- | -------------------- |
| **Range**              | Total spread                | Same as data | Simple            | Affected by outliers |
| **Variance**           | Average squared deviation   | Squared      | Used in formulas  | Harder to interpret  |
| **Standard Deviation** | Typical deviation from mean | Same as data | Easy to interpret | Affected by outliers |

## Skewness and Kurtosis — Understanding Distribution Shapes
Besides center and spread, it’s important to understand shape:
- Is the data symmetric or skewed?
- Are there heavy tails or many extreme values?

### Skewness
Skewness measures asymmetry of the distribution.
```
skew_income = df["income"].skew()
print(f"Skewness (income): {skew_income:.2f}")
```

Interpretation:
- Skew = 0 → perfectly symmetric
- Skew > 0 → right-skewed (long tail on the right; e.g., income data)
- Skew < 0 → left-skewed (long tail on the left)

Example:
- Income or sales data are often right-skewed — most people earn around the average, but a few earn much more.

````python
import pandas as pd

# Load the dataset
df = pd.read_csv('/content/drive/MyDrive/POSI/Lecture 2/Data/incomes_for_dispersion.csv')

print("🔍 Dataset Preview:")
print(df.head())

print("\n📊 Summary Statistics:")
print(df["income"].describe())
````

Interpretation:

- The red KDE line (Kernel Density Estimate) smooths the curve.

| Metric                           | Meaning                     | Interpretation                                                                                                             |
| -------------------------------- | --------------------------- | -------------------------------------------------------------------------------------------------------------------------- |
| **Range (85,000)**               | Max − Min                   | Shows total spread between the smallest and largest income. Indicates the dataset has large variability (due to outliers). |

Interpretation:

- The green lines show one standard deviation above and below it.

## Kurtosis
Kurtosis measures how peaked or flat the distribution is.

```
kurt_income = df["income"].kurtosis()
print(f"Kurtosis (income): {kurt_income:.2f}")
```

Interpretation:
- Kurtosis ≈ 3 → normal distribution (mesokurtic)
- Kurtosis > 3 → heavy tails, sharp peak (leptokurtic)
- Kurtosis < 3 → flat, light tails (platykurtic)

This can indicate whether outliers are likely. For example, a high kurtosis means there are a few extreme values.

````python
import pandas as pd

# Calculate skewness and kurtosis for a specific column
skew_value = df["income"].skew()
kurtosis_value = df["income"].kurtosis()

print(f"Skewness: {skew_value:.2f}")
print(f"Kurtosis: {kurtosis_value:.2f}")
````

Explanation

df["income"].skew() → measures asymmetry of the data:
- 0 → symmetric
- > 0 → right-skewed (tail to the right)
- < 0 → left-skewed (tail to the left)

df["income"].kurtosis() → measures peakedness or heaviness of tails:
- Around 3 → normal (mesokurtic)
- > 3 → peaked, heavy tails (leptokurtic)
- < 3 → flat, light tails (platykurtic)

# Univariate Analysis

## What Is Univariate Analysis?

Univariate analysis means analyzing one variable (column) independently to understand:
- What values it takes,
- How they are distributed,
- Whether there are outliers or missing values,
- And what patterns or trends might exist.

## Tools Usefull For Univariate Analysis

For numerical data:
- Histograms
- Boxplots
- Density plots

For categorical variables:
- bar charts
- frequency tables
- pie charts

````python
# ======================================
# 1️⃣ Import Libraries and Load Datasets
# ======================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load datasets (adjust path if needed)
num_df = pd.read_csv("/content/drive/MyDrive/POSI/Lecture 2/Data/numerical_dataset.csv")
cat_df = pd.read_csv("/content/drive/MyDrive/POSI/Lecture 2/Data/categorical_dataset.csv")

# Display first few rows
print("Numerical dataset:")
print(num_df.head(), "\n")

print("Categorical dataset:")
print(cat_df.head())
````

````python
# --------------------------------------
# Histogram for numerical data
# --------------------------------------
plt.figure(figsize=(8, 5))
plt.hist(num_df["income"], bins=15, color='skyblue', edgecolor='black')
plt.title("Histogram of Income")
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


# --------------------------------------
# Boxplot for numerical data
# --------------------------------------
plt.figure(figsize=(7, 3))
plt.boxplot(num_df["income"], vert=False, patch_artist=True,
            boxprops=dict(facecolor='lightgreen', color='black'))
plt.title("Boxplot of Income")
plt.xlabel("Income")
plt.show()


# --------------------------------------
# Density plot (KDE) for numerical data
# --------------------------------------
plt.figure(figsize=(8, 5))
sns.kdeplot(num_df["income"], fill=True, color='orange', alpha=0.6)
plt.title("Density Plot of Income")
plt.xlabel("Income")
plt.ylabel("Density")
plt.show()
````

````python
# --------------------------------------
# Frequency table for categorical data
# --------------------------------------
print("Frequency Table for Fruits:")
freq_table = cat_df["fruit"].value_counts()
print(freq_table, "\n")


# --------------------------------------
# Bar chart for categorical data
# --------------------------------------
plt.figure(figsize=(7, 5))
freq_table.plot(kind='bar', color='lightcoral', edgecolor='black')
plt.title("Bar Chart of Favorite Fruits")
plt.xlabel("Fruit")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()


# --------------------------------------
# Pie chart for categorical data
# --------------------------------------
plt.figure(figsize=(6, 6))
freq_table.plot(kind='pie',
                autopct='%1.1f%%',
                startangle=90,
                colors=sns.color_palette("pastel"),
                ylabel='')
plt.title("Pie Chart of Favorite Fruits")
plt.show()
````

````python
fig, axes = plt.subplots(2, 3, figsize=(15, 8))

# Histogram
axes[0, 0].hist(num_df["income"], bins=15, color='skyblue', edgecolor='black')
axes[0, 0].set_title("Histogram of Income")

# Boxplot
axes[0, 1].boxplot(num_df["income"], vert=False, patch_artist=True,
                   boxprops=dict(facecolor='lightgreen', color='black'))
axes[0, 1].set_title("Boxplot of Income")

# Density
sns.kdeplot(num_df["income"], fill=True, color='orange', ax=axes[0, 2])
axes[0, 2].set_title("Density Plot of Income")

# Frequency Table (Bar)
freq_table.plot(kind='bar', ax=axes[1, 0], color='lightcoral', edgecolor='black')
axes[1, 0].set_title("Bar Chart of Fruits")

# Pie Chart
freq_table.plot(kind='pie', ax=axes[1, 1], autopct='%1.1f%%',
                startangle=90, colors=sns.color_palette("pastel"), ylabel='')
axes[1, 1].set_title("Pie Chart of Fruits")

# Hide empty subplot
axes[1, 2].axis("off")

plt.tight_layout()
plt.show()
````
